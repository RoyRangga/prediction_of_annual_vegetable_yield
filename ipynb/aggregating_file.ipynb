#### inspeksi nilai -

```python
#disini kuat diduga bahwa tanda "-" merepresentasikan bahwa tidak ada data direcord, itu artinya "-" dapat diasumsikan sebagai nan
```

```python
inspect = sayur.copy()
```

```python
inspect[inspect.province_name == "KALIMANTAN UTARA"]
```

```python
df = pd.DataFrame()
df["col_name"] = inspect.columns.tolist()
df["num_nan"] = [(inspect[f'{i}'] == "-").sum() for i in inspect.columns.tolist()]
df["all_rows"] = inspect.shape[0]
df["percentage_of_nan"] = ((df["num_nan"]/df["all_rows"])*100)
df
#(inspect[f'{i}']=="-").sum()
```

```python
inspect.info()
```

```python
for i in inspect.columns.tolist():
    inspect[f"{i}"] = inspect[f"{i}"].replace("-", np.nan)
```

```python
for i in inspect.columns.tolist()[1:-1]:
    inspect[f"{i}"] = inspect[f"{i}"].astype(float)
```

```python
inspect.info()
```

```python
#mengisi missing values berdasarkan provinsi tertentu
for i in inspect.columns.tolist()[1:-1]:
    grouped_province_mean = inspect.groupby("province_name")[f"{i}"].transform("mean")
    inspect[f"{i}"].fillna(grouped_province_mean, inplace=True)
```
